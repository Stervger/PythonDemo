```python
df.columns = df.columns.map(lambda x:x.upper())
```

```python
dff = df.loc[:,['LOAD_TIME','FFP_DATE','LAST_TO_END','FLIGHT_COUNT','SEG_KM_SUM','AVG_DISCOUNT']]
```

```python
dff.describe()
```

```python
dff['L'] = dff['LOAD_TIME'] - dff['FFP_DATE']
```

```python
dff.head()
```

```python
dff['R'] = dff.LAST_TO_END
```

```python
dff['F'] = dff.FLIGHT_COUNT
```

```python
dff['M'] = dff.SEG_KM_SUM
```

```python
dff['C'] = dff.AVG_DISCOUNT
```

```python
dff['L'] = dff.L.map(lambda x:x.days)
```

```python
data = dff.iloc[:,-5:]
```

```python
data
```

```python
data['L'] = data.L/30
```

```python
data
```

```python
data.describe()
```

```python
data.R = data.R/30
```

```python
data
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
zdata = std.fit_transform(data)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 5)
```

```python
model.fit(zdata)
```

```python
model.labels_
```

```python
zdata
```

```python
df_z = pd.DataFrame(zdata, index = df.index, columns = ['ZL','ZR','ZF','ZM','ZC'])
```

```python
df_z.head()
```

```python
df_z['labels'] = model.labels_
```

```python
df_z.head()
```

```python
df1 = df_z[df_z.labels == 0]
df2 = df_z[df_z.labels == 1]
df3 = df_z[df_z.labels == 2]
df4 = df_z[df_z.labels == 3]
df5 = df_z[df_z.labels == 4]
```

```python
mean1 = df1.describe().loc['mean',['ZL','ZR','ZF','ZM','ZC']]
mean2 = df2.describe().loc['mean',['ZL','ZR','ZF','ZM','ZC']]
mean3 = df3.describe().loc['mean',['ZL','ZR','ZF','ZM','ZC']]
mean4 = df4.describe().loc['mean',['ZL','ZR','ZF','ZM','ZC']]
mean5 = df5.describe().loc['mean',['ZL','ZR','ZF','ZM','ZC']]
```

```python
mean1
```

```python
angles = np.linspace(0,2*np.pi,5,endpoint = False)
```

```python
angles
```

```python
ax = plt.subplot(111,polar = True)
ax.plot(angles, mean1,'ro--',label = 'class1')
ax.plot(angles, mean2,'bo--',label = 'class2')
ax.plot(angles, mean3,'yo--',label = 'class3')
ax.plot(angles, mean4,'go--',label = 'class4')
ax.plot(angles, mean5,'ko--',label = 'class5')
ax.set_thetagrids(angles * 180/np.pi, ['ZL','ZR','ZF','ZM','ZC'])
plt.legend(loc='lower right', bbox_to_anchor=(1.5, 0.0)) 
```
