```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
df = pd.read_excel('C:/Users/Administrator/Desktop/pandas_practise/consumption_data.xls',index_col = 0);
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
data = std.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3, max_iter = 500, n_jobs = 4)
```

```python
model.fit(data)
```

```python
model.labels_
```

```python
df['labels'] = model.labels_
```

```python
df
```

```python
df_1 = df[df['labels'] == 0]
df_2 = df[df['labels'] == 1]
df_3 = df[df['labels'] == 2]
```

```python
df_3.describe()
```

```python
df_1.iloc[:,:-1].plot(kind = 'kde',subplots = True,sharex = False)
plt.show()
```

```python
df_2.iloc[:,:-1].plot(kind = 'kde',subplots = True,sharex = False)
plt.show()
```

```python
df_3.iloc[:,:-1].plot(kind = 'kde',subplots = True,sharex = False)
plt.show()
```
